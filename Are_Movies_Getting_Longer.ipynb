```python
netflix_df = pd.read_csv("/content/gdrive/MyDrive/Python-Practice/netflix_data.csv")
netflix_df.head()
```

```python
netflix_df_movies_only = netflix_df[netflix_df["type"] == 'Movie']
netflix_movies_col_subset = netflix_df_movies_only[["title", "country", "genre", "release_year", "duration"]]
netflix_movies_col_subset.head()
```

```python
netflix_movies_col_subset.plot.scatter(x = "release_year", y = "duration");
plt.title("Movie Duration by Year of Release");
```

```python
movie_length = netflix_movies_col_subset.groupby('release_year', as_index= False)['duration'].mean()
movie_length.head()
```

```python
plt.plot(movie_length['release_year'],movie_length['duration']);
```

```python
short_movies = netflix_movies_col_subset[netflix_movies_col_subset["duration"] < 60]
short_movies.head(20)
```

```python
colors = []
for lab, row in netflix_movies_col_subset.iterrows():
    if row['genre'] == "Children":
        colors.append("red")
    elif row['genre'] == "Documentaries":
        colors.append("blue")
    elif row['genre'] == "Stand-Up":
        colors.append("green")
    else:
        colors.append("black")
colors[0:10]
```

```python
plt.style.use('fast')
fig = plt.figure(figsize=(12,8))
```

```python
plt.scatter(netflix_movies_col_subset["release_year"], netflix_movies_col_subset["duration"], color = colors);
plt.title("Movie duration by year of release");
plt.xlabel("Release year");
plt.ylabel("Duration (min)");
```
